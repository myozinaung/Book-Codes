```python
# import required libraries
import math as math              # for maths functions such as sqrt, log10, fmod
import numpy as np               # for array of state data
import matplotlib.pyplot as plt  # for plotting
import matplotlib.gridspec as gridspec
import time                      # for recording cpu time
# Own Libraries
import Filters
import Integrators
import PID
import random
from scipy import fft
from PID_MYO import PID_Object
from LQR_MYO import LQR
```

```python
#State Function
def stateFun(t,x,u,p):
    ### INPUTS ###
    # t     = time [for time-dependent terms]
    # x     = state vector
    # u     = control vector
    # p     = parameters [for cosnstant and time-varying]
    
    ### OUTPUT ###
    # x_dot = derivative of x (same length as x)
    x_dot = np.zeros(x.size)
    
    # ------Start of own code------
    K = 0.1705
    T = 7.1167
    
    r     = x[1]
    delta = u[0]
 
    x_dot[0] = r
    x_dot[1] = (K*delta-r)/T + 0.01
    # ------End of own code--------
    
    return x_dot
```

```python
# Helper Functions
# Clamp x between x_min and x_max
def clamp(x,x_min,x_max):
    if x <= x_min:
        return x_min
    elif x >= x_max:
        return x_max
    else:
        return x
```

```python
# Helper Functions
# Clapm x between x_min and x_max
def clamp(x,x_min,x_max):
    if x <= x_min:
        return x_min
    elif x >= x_max:
        return x_max
    else:
        return x

# Rewrite an angle(rad) within -pi to pi
def pi2pi(x):
    x = math.fmod(x+np.sign(x)*math.pi,2*math.pi)- np.sign(x)*math.pi
    # Note: Python x % y doesn't work same as math.fmod
    return x

```

```python
# Plot
def plot(t_vec,x_mat,u_mat, x_d_mat):
    fig = plt.figure()
#    gs = gridspec.GridSpec(2, 1, figure=fig)
    ax1 = fig.add_subplot(311)
    ax1.plot(t_vec, x_mat[0,:])
    ax1.plot(t_vec, x_d_mat[0,:])
    ax1.grid()
    
    ax2 = fig.add_subplot(312)
    ax2.plot(t_vec, x_mat[1,:])
    ax2.plot(t_vec, x_d_mat[1,:])
    ax2.grid()
    
    ax3 = fig.add_subplot(313)
    ax3.plot(t_vec, u_mat[0,:])
    ax3.grid()
    fig.savefig("test.png")
    plt.show()
```

```python
# MAIN PROGRAM
# Initial Conditions
psi0 = 0*math.pi/180
r0   = 0
x0   = np.array([psi0, r0])

delta0 = 0
u0 = np.array([delta0])

# Start, Finish, Time Step
t0 = 0
tF = 100
dt = 0.01

# PID Gains
KP = 3
KI = 0.05*1
KD = 10
PID_Gains = np.array([KP, KI, KD])


# # For PIDAutopilot
# KP = 10
# KI = 0.05
# TD = 50
# TE = 2
# PID_Gains = np.array([KP, KI, TD, TE]) # For PIDAutopilot

## For LQR
K = 0.1705
T = 7.1167
# A = np.array([[0, 1],
#               [0, -1/T]])
# B = np.array([[0],
#               [K/T]])
# C = np.array([[1, 0]])
# # Q = np.array([[1, 0],
# #               [0, 1]])
# Q = np.eye(1)
# R = np.array([[1]])

# For Integral LQR
A = np.array([[0, 1, 1],
              [0, 0, 1],
              [0, 0, -1/T]])
B = np.array([[0],
              [0],
              [K/T]])

Q = np.array([[0.5, 0, 0],
              [0, 1, 0],
              [0, 0, 1]])
R = np.array([[1]])
I_error = 0
              
delta_max = 35*math.pi/180
delta_rate_max = 60*math.pi/180
u_bound = np.array([-delta_max, delta_max])
u_rate_bound = np.array([-delta_rate_max, delta_rate_max])

######################## Simulate ########################
#--------------------------------------------------------#
t_start = time.time() # to record cpuTime
print("Simulation Start!")
psi_ref = 45*math.pi/180
x_ref = np.array([psi_ref, 0])

x = x0
u = u0
p = []
error_last = 0
Integral   = 0
x_filter   = 0
omega_f    = 2

# For Reference Generator
# x_d = np.array([psi0, r0]) # For 2nd order RefGen
x_d = np.array([psi0, r0, 0]) # For 3rd order RefGen
omega_n = 0.5  # How fast the psi_d converged to psi_ref
zeta    = 1.5
r_d_max = 0.08

K_add = 0 # Added Inertia in Feedforward Term (0 to 1)

t_vec = np.arange(t0,tF,dt)
x_mat = np.zeros((x0.size, t_vec.size))
x_d_mat = np.zeros((x0.size+1, t_vec.size))
u_mat = np.zeros((u0.size, t_vec.size))

x_PID = np.array([[0],
                  [0]]) # Filtered PID
N = 1 # filtered coefficient (should tune together with PID gains)

Noise_Amp = 0.05*0
myPID = PID_Object(KP, KI, KD, u_bound)
# myLQR = LQR(A, B, C.T@Q@C, R, u_bound)
myLQR = LQR(A, B, Q, R, u_bound)
idx = 0
for t in t_vec:

    psi = x[0] + (random.uniform(-1,1))*Noise_Amp
#     psi, x_filter = Filters.Lowpass4th(x_filter, psi, omega_f, dt)
    psi = pi2pi(psi) # Rewrite with -pi and pi
    psi_ref = pi2pi(psi_ref)
    error = psi_ref - psi
    error = pi2pi(error)
    r = x[1]
#        u = PID.P(error, PID_Gains, u_bound)     # Select Control Method
#        u = PID.PD(error, error_last, dt, PID_Gains, u_bound)     # Select Control Method
#     u, Integral = PID.PID(error, error_last, Integral, dt, PID_Gains, u_bound)     # Select Control Method
#     u, x_PID = PID.FilteredPID(x_PID, error, dt, PID_Gains, N, u_bound)
#     u, Integral = PID.PIDAutopilot(error, u, r, Integral, dt, PID_Gains, u_bound, u_rate_bound)
    
#     u = myPID(error, dt)
    x[0] = psi
#     x_d = Filters.RefGen2nd(x_d, psi_ref, omega_n, zeta, dt, r_d_max)
    x_d = Filters.RefGen3rd(x_d, psi_ref, omega_n, zeta, dt, r_d_max)
#     I_error += 0.01*(error)*dt
    I_error = x[0] - x_d[0]
    error_vec = x - x_d[:2]
    
    u, LQR_Gain = myLQR(np.append(I_error,error_vec)) 
#     u, LQR_Gain = myLQR(error_vec) 

#     u_FF = Filters.Nomoto_FF(x_d, K_add) # Need 3rd order Ref Gen
#     u = u + u_FF

#     u = u + I_error
    u = clamp(u, u_bound[0], u_bound[1])
    error_last = error
    u = np.array([u])

#    x = euler(stateFun,t,x,u,p,dt)   # Select Integration Method
    x = Integrators.RKGill(stateFun,t,x,u,p,dt)   # Select Integration Method

    # Save State and Control Data
    x_mat[:,idx] = x
    u_mat[:,idx] = u
    x_d_mat[:,idx] = x_d

    idx += 1
del myPID
print("Simulation Finished!")
t_end   = time.time() # to record cpuTime
#--------------------------------------------------------#
##########################################################

cpuTime = round(t_end - t_start,2)
print("cpuTime:",cpuTime, "sec")

print(x_mat[0,-1]*180/math.pi)
plot(t_vec,x_mat,u_mat, x_d_mat) # call plot function
```
